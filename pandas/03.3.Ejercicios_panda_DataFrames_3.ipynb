2. Escribe un programa Pandas para contar la supervivencia por género, categorías y edad de varias clases. Siendo las categorías de edad: (0, 10), (10, 30), (30, 60), (60, 80)
Agrupar por categoría de edad y contemos por genero y categoría los supervivientes

```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('/home/azureuser/cloudfiles/code/Users/belenolmosvalverde24/belendata/Pandas/titanic.csv')

# Definir los límites de las categorías de edad. Aquí definimos los límites (bins) para las categorías de edad y las etiquetas (labels) que se asignarán a cada categoría.
bins = [0, 10, 30, 60, 80]
labels = ['0-10', '10-30', '30-60', '60-80']

# Crear una nueva columna con las categorías de edad.Utilizamos pd.cut para dividir la columna age en las categorías definidas por bins 
#y asignamos las etiquetas correspondientes. La opción right=False asegura que el límite superior de cada bin no se incluya en la categoría.
#Utilizamos pd.cut para dividir la columna age en las categorías definidas por bins y asignamos las etiquetas correspondientes. 
#La opción right=False asegura que el límite superior de cada bin no se incluya en la categoría.
df['age_category'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Agrupar por género, categoría de edad y clase, y contar las supervivencias
#df.groupby(['sex', 'age_category', 'pclass']): Agrupa los datos por las columnas sex (género), age_category (categoría de edad) y pclass (clase).
#['survived'].sum(): Dentro de cada grupo, suma los valores de la columna survived para contar cuántas personas sobrevivieron.
#reset_index(): Convierte el resultado agrupado en un DataFrame plano, restableciendo los índices.

survival_counts = df.groupby(['sex', 'age_category', 'pclass'])['survived'].sum().reset_index()

# Mostrar el resultado
print(survival_counts)
```

```python
#Solución
pcdf = df.pivot(columns='class', values='survived')
print(pcdf)
ndf = pd.concat([df,pcdf], axis=1)
ndf

```

```python
#Solución age es age_category*
ndf.groupby(['sex','age_category']).agg({'First':'count','Second':'count', 'Third':'count'})
```

```python
#Solución con pivot_table
df.pivot_table('survived', index=['sex', 'age_category'], columns='class', aggfunc='count')
```

3. Escribe un programa Pandas para encontrar la tasa de supervivencia por género y edad de las diferentes categorías de varias clases. 
Agregar la tarifa como una dimensión de columnas y dividir la columna de tarifa en 2 categorías según los valores presentes en las columnas de tarifa.

```python
import pandas as pd

# Cargar el conjunto de datos Titanic desde el archivo CSV
df = pd.read_csv('/home/azureuser/cloudfiles/code/Users/belenolmosvalverde24/belendata/Pandas/titanic.csv')

# Crear una nueva columna 'Fare_Category' dividiendo la tarifa en 2 categorías: low fare y high fare
df['Fare_Category'] = pd.qcut(df['fare'], 2, labels=['Low Fare', 'High Fare'])

# Crear una nueva columna 'Age_Category' dividiendo la edad en 2 categorías: child y adult
df['Age_Category'] = pd.cut(df['age'], bins=[0, 18, 80], labels=['Child', 'Adult'])

# Calcular la tasa de supervivencia por género, edad, clase y categoría de tarifa utilizando una tabla dinámica
survival_rate = df.pivot_table('survived', index=['sex', 'Age_Category'], columns=['pclass', 'Fare_Category'], aggfunc='mean')

print(survival_rate)
```

5. Escribe un programa Pandas para separar el género según si viajaron solos o no y obtener la probabilidad de supervivencia.

```python
import pandas as pd

#Carga el conjunto de datos Titanic desde el archivo CSV.
df = pd.read_csv('/home/azureuser/cloudfiles/code/Users/belenolmosvalverde24/belendata/Pandas/titanic.csv')

# Crea una nueva columna Alone que indica si la persona viajó sola (sin hermanos/esposos ni padres/hijos).
df['Alone'] = (df['sibsp'] == 0) & (df['parch'] == 0)

# Calcula la probabilidad de supervivencia por género y si viajaron solos o no utilizando una tabla dinámica.

#df.pivot_table`: Esta función de Pandas crea una tabla dinámica (pivot table) a partir del DataFrame `df`.

#survived': Este es el valor que queremos agregar en la tabla dinámica. En este caso, estamos interesados en la columna `survived`, que indica si un pasajero sobrevivió (1) o no (0).

#index='sex': Esto especifica que queremos que las filas de la tabla dinámica estén indexadas por el género (`sex`). Así, tendremos una fila para `male` y otra para `female`.

#columns='Alone': Esto especifica que queremos que las columnas de la tabla dinámica estén divididas según si el pasajero viajó solo (`Alone` es `True`) o no (`Alone` es `False`).

#aggfunc='mean': Esta es la función de agregación que queremos aplicar. En este caso, `mean` calcula el promedio de los valores en la columna `survived` para cada combinación de género y si viajaron solos o no. Dado que los valores en `survived` son 0 o 1, el promedio de estos valores nos da la probabilidad de supervivencia.

survival_probability = df.pivot_table('survived', index='sex', columns='Alone', aggfunc='mean')

print(survival_probability)
```

6. Escribe un programa Pandas para crear encontrar la probabilidad de supervivencia por clase, género, embarque en solitario y puerto de embarque.

```python
import pandas as pd

# Cargar el conjunto de datos Titanic desde el archivo CSV
df = pd.read_csv('/home/azureuser/cloudfiles/code/Users/belenolmosvalverde24/belendata/Pandas/titanic.csv')

# Crea una nueva columna Alone que indica si la persona viajó sola (sin hermanos/esposos ni padres/hijos).
df['Alone'] = (df['sibsp'] == 0) & (df['parch'] == 0)

# Calcula la probabilidad de supervivencia por clase (pclass), género (sex), si viajaron solos (Alone) y puerto de embarque (embarked) utilizando una tabla dinámica (pivot_table).
survival_probability = df.pivot_table('survived', index=['pclass', 'sex', 'Alone', 'embarked'], aggfunc='mean')

print(survival_probability)
```
