```python
import statistics #Importa el módulo statistics que proporciona funciones para realizar cálculos estadísticos (media y desviación estándar en este caso).

def generate_new_data(filename): #Define la función generate_new_data que toma el nombre de un archivo como argumento.
    if not os.path.isfile(filename): #if not os.path.isfile(filename): verifica si el archivo existe. Si no existe, la función devuelve None.
        return None
    fobj_in = open(filename) #Abre el archivo en modo lectura y almacena el objeto de archivo en fobj_in.
    idx = 0 #Inicializa un índice para seguir la posición de las filas.
    new_data = [] #Inicializa una lista vacía para almacenar los datos procesados.
    for line in fobj_in: #recorre cada línea del archivo.
        alumno = line.rstrip().split(',') #Elimina los espacios en blanco al final de la línea y divide la línea en una lista de valores usando la coma como separador.
        new_line = list(alumno) #Copia la lista de valores en new_line.
        mean ='media' #Inicializa la media.
        stdv ='stdv'#Inicializa la desviación estándar.

        if idx > 0: #para saltarse la cabecera, se procede a calcular la media y desviación estándar.
            notas = (int(alumno[1]), int(alumno[2])) #Convierte las notas de texto a enteros y las almacena en una tupla.
            mean = statistics.mean(notas) #Calcula la media de las notas.
            stdv = statistics.stdev(notas) #Calcula la desviación estándar de las notas.

        new_line.append(mean)#Añade la media estándar calculadas a la lista new_line
        new_line.append(stdv)#Añade la desviación estándar calculadas a la lista new_line

        new_data.append(new_line)#Añade new_line a la lista new_data
        idx += 1 #Incrementa el índice idx.

    fobj_in.close() #Cierra el archivo de entrada.

    return new_data #Devuelve la lista new_data con los datos procesados.


def write_new_data(filename, aList): #Define la función write_new_data que toma un nombre de archivo y una lista de datos como argumentos.
    if aList and len(aList) > 0: #verifica si la lista no está vacía.
        with open(filename, 'w') as f: #Abre el archivo en modo escritura ('w') utilizando una sentencia with, que asegura que el archivo se cierra automáticamente al final del bloque.
            for line in aList:#recorre cada línea de datos en la lista.
                linea = ','.join(map(lambda l: str(l), line)) #Convierte cada elemento de la línea a una cadena y los une con comas.
                f.write(linea + '\n') #Escribe la línea en el archivo.
            f.close() #Aunque with ya cierra el archivo automáticamente, se incluye f.close() explícitamente (es redundante aquí).
    else:
        print('Lista vacia') #Si la lista está vacía, imprime un mensaje indicando que la lista está vacía.


data = generate_new_data(raiz + "../data/notas.csv") #Llama a la función generate_new_data con la ruta del archivo CSV original y almacena los datos procesados en data.
print(data) #Imprime los datos procesados.

write_new_data(raiz + '../data/notas_new.csv', data) #para escribir los datos procesados en un nuevo archivo CSV.
```
